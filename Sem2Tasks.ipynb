## Задание 2. Составьте шаблон дизайна эксперимента для гипотезы, которая набрала больше всего баллов.  

Больше всего баллов набрала **_Третья гипотеза_**

**Шаблон дизайна эксперимента:**
1. _Гипотеза_  
Если на сайте: https://www.citilink.ru/ разработать и интегрировать пользовательские функции,
такие как фильтры поиска, порекомендованные товары, часто покупаемые, то увеличатся
показатели Daily Active Users и Retention Rate, это даст возможность пользователям лучше
контролировать свои финансы, а для организации, увеличить число продаж.
2. _Что делаем:_  
• Контрольная версия: оставляем сайт в текущем виде.  
• Тестовая версия: добавляем на сайт, разработанные и интегрированные пользовательские
функции, описанные в гипотезе.  
3. _На каких пользователях тестируем:_  
Только на новых пользователях.
4. _Ключевые метрики для оценки эксперимента:_  
• Daily Active Users – число активных пользователей, которые взаимодействуют
с приложением каждый день.  
• Retention Rate — коэффициент удержания клиентов, показывающий сколько
пользователей возвращается в мобильное приложение.  
5. _Ожидаемый эффект_  
Увеличение показателя Daily Active Users на 5%, Retention Rate – на 10%
6. _План действий в зависимости от результатов эксперимента_  
• Если результат эксперимента будет положительным и прирост значений
ключевых метрик будет не менее ожидаемого, то распространяем изменения на всех
пользователей.  
• Если значения ключевых метрик снижаются или имеют незначительный
прирост (менее ожидаемого), то откатываем эксперимент.
